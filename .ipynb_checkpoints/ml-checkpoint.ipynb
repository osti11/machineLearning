# Importanweisung
## Allgemein

```python
from __future__ import print_function # print kann jetzt in älteren Verionen als Pyton 3.x als Funktion verwendet werden, normal nur ANweisung

from time import time # importiert die Funktion 'time()' aus dem Modul 'time'

import numpy as np # impotiert das Modul 'numpy', Zum erstell von mehrdimensionalen Array's und Werkzeuge um diese zu bearbeiten
import matplotlib.pyplot as plt # importiert das Modul 'matplotlib.pyplot', zum Erstellen von Figuren und Achsen

from sklearn.datasets.base import Bunch # lädt die Funktion 'sklearn.datasets.base.bunch()' zum managen der Daten 
from sklearn.datasets import fetch_species_distributions # Lädt den Artenverteilungsdatensatz von Phillips et. al. (2006) als Bunch Objekt
from sklearn.datasets.species_distributions import construct_grids
from sklearn import svm, metrics # importiert die Module svm und metrics
# svm:     Support Vector Machine-Algorithmen
# metrics: Bewertungsfunktionen, Leistungsmetriken sowie paarweise Metriken und Abstandsberechnungen
```

## Grundkarte, für die Landesgrenzen von Südafrika

```python
try: # Basemap wird verwendet wenn verfügbar.
    from mpl_toolkits.basemap import Basemap # impotiert das 'matplotlib basemap toolkit' 
    basemap = True
except ImportError:
    basemap = False
    
print("basemap: " + str(basemap))
```

# Daten aufbereiten

```python
print(__doc__) # Ausgabe des Docstring

def create_species_bunch(species_name, train, test, coverages, xgrid, ygrid): # Beschreibung
    """Create a bunch with information about a particular organism

    This will use the test/train record arrays to extract the
    data specific to the given species name.
    """
    
    bunch = Bunch(name=' '.join(species_name.split("_")[:2])) # Ersetzt "_" durch " " im Namen der Spezies
    species_name = species_name.encode('ascii') # kodiert Namen in ASCII
    points = dict(test=test, train=train) # Erstellt jeweils ein Dictionary für die Test- und Trainingsdaten
    
    for label, pts in points.items():
        pts = pts[pts['species'] == species_name] # Wählt die Punkte aus, die der gewünschten Art zugeordnet sind
        bunch['pts_%s' % label] = pts

        # Ermitteln Sie die Erfassungswerte für jeden Trainings- und Testpunkt
        ix = np.searchsorted(xgrid, pts['dd long'])
        iy = np.searchsorted(ygrid, pts['dd lat'])
        bunch['cov_%s' % label] = coverages[:, -iy, ix].T

    return bunch
```

# Trainieren, Vorhersagen und grafische Darstellung

```python
def plot_species_distribution(species=("bradypus_variegatus_0",
                                       "microryzomys_minutus_0")):

    if len(species) > 2:
        print("Note: when more than two species are provided,"
              " only the first two will be used")

    t0 = time() 

    # Lädt die komprommierten Daten
    data = fetch_species_distributions()

    # Datenraster einrichten
    xgrid, ygrid = construct_grids(data)

    # Das Raster in x,y Kordinaten 
    X, Y = np.meshgrid(xgrid, ygrid[::-1])

    # Erstellt ein 'bunch' Objekt für jede Spezies
    BV_bunch = create_species_bunch(species[0],
                                    data.train, data.test,
                                    data.coverages, xgrid, ygrid)
    MM_bunch = create_species_bunch(species[1],
                                    data.train, data.test,
                                    data.coverages, xgrid, ygrid)

    # Grid Kordinaten für die Auswertung
    np.random.seed(13)
    # Zufaellige Zahlen zwischen 0 und der Anzahl der Koordinaten in eine Richtung
    background_points = np.c_[np.random.randint(low=0, high=data.Ny,
                                                size=10000),
                              np.random.randint(low=0, high=data.Nx,
                                                size=10000)].T

    # data.coverages[6] = Wo Messungen durchgeführt wurden
    # keine Mesungen = Wasser
    # Festlegn von Land und Wasser für die Darstellung
    land_reference = data.coverages[6]

    # Trainieren, Vorhersagen und grafische Darstellung für jede Art.
    for i, species in enumerate([BV_bunch, MM_bunch]):
        print("_" * 80)
        print("Modellierung der Artenverteilung '%s'" % species.name)

        # Standardize features
        mean = species.cov_train.mean(axis=0) # Durchschnittswert der Umweltfaktoren
        std = species.cov_train.std(axis=0) # Standardabweichung der Umweltfaktoren
        train_cover_std = (species.cov_train - mean) / std
        
        # Trainieren des Classifiers 'OneClassSVM'
        print(" - trainiere OneClassSVM ... ", end='')
        clf = svm.OneClassSVM(nu=0.1, kernel="rbf", gamma=0.5)
        clf.fit(train_cover_std)
        print("fertig.")

        # Plot Karte von Süd Afrika
        plt.subplot(1, 2, i + 1)
        if basemap:
            print(" - plot coastlines using basemap")
            m = Basemap(projection='cyl', llcrnrlat=Y.min(),
                        urcrnrlat=Y.max(), llcrnrlon=X.min(),
                        urcrnrlon=X.max(), resolution='c')
            m.drawcoastlines()
            m.drawcountries()
        else:
            print(" - plot coastlines from coverage")
            plt.contour(X, Y, land_reference,
                        levels=[-9998], colors="k",
                        linestyles="solid")
            plt.xticks([])
            plt.yticks([])

        print(" - Vorhersage der Artenverteilung")

        # Vorhersage der Artenverteilung anhand der Trainingsdaten
        Z = np.ones((data.Ny, data.Nx), dtype=np.float64)

        # Vorhersage nur für das Land.
        idx = np.where(land_reference > -9999)
        coverages_land = data.coverages[:, idx[0], idx[1]].T

        pred = clf.decision_function((coverages_land - mean) / std)
        Z *= pred.min()
        Z[idx[0], idx[1]] = pred
        
        # Abstufungen in regelmaessigen Abstaenden zw. dem kleinsten und groessten Wert, hier 25 Schritte
        levels = np.linspace(Z.min(), Z.max(), 25)
        Z[land_reference == -9999] = -9999

        # plot contours of the prediction
        plt.contourf(X, Y, Z, levels=levels, cmap=plt.cm.Reds)
        plt.colorbar(format='%.2f')

        # Trainings-/TeststreuPunkte 
        plt.scatter(species.pts_train['dd long'], species.pts_train['dd lat'],
                    s=2 ** 2, c='blue',
                    marker='^', label='train')
        plt.scatter(species.pts_test['dd long'], species.pts_test['dd lat'],
                    s=2 ** 2, c='green',
                    marker='x', label='test')
        plt.legend()
        plt.title(species.name)
        plt.axis('equal')

        # Compute AUC (Area Under Curve) with regards to background pointss
        pred_background = Z[background_points[0], background_points[1]]
        pred_test = clf.decision_function((species.cov_test - mean) / std)
        scores = np.r_[pred_test, pred_background]
        y = np.r_[np.ones(pred_test.shape), np.zeros(pred_background.shape)]
        fpr, tpr, thresholds = metrics.roc_curve(y, scores)
        roc_auc = metrics.auc(fpr, tpr)
        plt.text(-35, -70, "AUC: %.3f" % roc_auc, ha="right")
        print("\n Area under the ROC curve : %f" % roc_auc)

    print("\ntime elapsed: %.2fs" % (time() - t0))


plot_species_distribution()
plt.show()
```
